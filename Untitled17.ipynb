```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
import base64, cv2, numpy as np

def capture_frame():
    js = Javascript("""
    async function takePhoto() {
      const video = document.createElement('video');
      document.body.appendChild(video);
      const stream = await navigator.mediaDevices.getUserMedia({video: true});
      video.srcObject = stream;
      await video.play();

      const canvas = document.createElement('canvas');
      canvas.width = video.videoWidth;
      canvas.height = video.videoHeight;
      const ctx = canvas.getContext('2d');
      ctx.drawImage(video, 0, 0);

      stream.getVideoTracks()[0].stop();
      video.remove();

      return canvas.toDataURL('image/jpeg', 0.95);
    }
    takePhoto();
    """)

    display(js)
    data = eval_js("takePhoto()")
    _, encoded = data.split(",", 1)
    img = base64.b64decode(encoded)
    arr = np.frombuffer(img, np.uint8)
    frame = cv2.imdecode(arr, cv2.IMREAD_COLOR)
    return frame
```

```python
import mediapipe as mp

mp_hands = mp.solutions.hands
mp_draw = mp.solutions.drawing_utils

hands = mp_hands.Hands(
    min_detection_confidence=0.5,
    min_tracking_confidence=0.5
)
```

```python
from google.colab.patches import cv2_imshow

while True:
    frame = capture_frame()
    if frame is None:
        print("No frame captured!")
        continue

    rgb = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
    result = hands.process(rgb)

    if result.multi_hand_landmarks:
        for hand_landmarks in result.multi_hand_landmarks:
            mp_draw.draw_landmarks(
                frame,
                hand_landmarks,
                mp_hands.HAND_CONNECTIONS
            )

    cv2_imshow(frame)

    # Stop using the stop button in Colab
    break
```

```python
# Video-processing version of your script (works in Colab & locally)
import cv2
import numpy as np
import time
import os
from pathlib import Path

# -----------------------
# CONFIG (tune these)
# -----------------------
VIDEO_PATH = "video.mp4"   # replaced below by uploaded path when in Colab
OUTPUT_PATH = "output_hand_proximity.mp4"

FRAME_WIDTH = 640           # processing width (frames will be resized to this)
FRAME_HEIGHT = 480

SKIN_LOWER = np.array([0, 30, 60])   # HSV lower bound for skin
SKIN_UPPER = np.array([20, 255, 255])# HSV upper bound for skin
MIN_HAND_AREA = 2000        # minimum contour area to be considered a hand

DOT_REL_POS = (0.85, 0.35)  # dot position as fraction of (width, height)
DOT_RADIUS = 18

SAFE_DIST = 220             # pixel thresholds (tweak)
WARNING_DIST = 120

FONT = cv2.FONT_HERSHEY_SIMPLEX
FONT_SCALE = 0.8
FONT_THICK = 2

COLOR_SAFE = (50, 200, 50)
COLOR_WARNING = (0, 200, 255)
COLOR_DANGER = (0, 0, 255)

# If running in Colab and you uploaded a file, replace VIDEO_PATH variable:
try:
    # If video_path variable from the upload cell exists, use it
    if 'video_path' in globals():
        VIDEO_PATH = video_path
except Exception:
    pass

def get_hand_centroid_from_frame(frame_bgr):
    hsv = cv2.cvtColor(frame_bgr, cv2.COLOR_BGR2HSV)
    mask = cv2.inRange(hsv, SKIN_LOWER, SKIN_UPPER)
    kernel = cv2.getStructuringElement(cv2.MORPH_ELLIPSE, (5,5))
    mask = cv2.medianBlur(mask, 5)
    mask = cv2.morphologyEx(mask, cv2.MORPH_CLOSE, kernel, iterations=2)
    mask = cv2.morphologyEx(mask, cv2.MORPH_OPEN, kernel, iterations=1)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    if not contours:
        return None, mask
    largest = max(contours, key=cv2.contourArea)
    area = cv2.contourArea(largest)
    if area < MIN_HAND_AREA:
        return None, mask
    M = cv2.moments(largest)
    if M["m00"] == 0:
        return None, mask
    cx = int(M["m10"] / M["m00"])
    cy = int(M["m01"] / M["m00"])
    return (cx, cy, int(area), largest), mask

def compute_state(distance):
    if distance <= WARNING_DIST:
        return "DANGER", COLOR_DANGER
    elif distance <= SAFE_DIST:
        return "WARNING", COLOR_WARNING
    else:
        return "SAFE", COLOR_SAFE

def process_video(in_path, out_path, target_w=FRAME_WIDTH, target_h=FRAME_HEIGHT):
    cap = cv2.VideoCapture(in_path)
    if not cap.isOpened():
        raise RuntimeError(f"Cannot open video: {in_path}")

    # read source FPS & frame count
    src_fps = cap.get(cv2.CAP_PROP_FPS) or 25.0
    frame_count = int(cap.get(cv2.CAP_PROP_FRAME_COUNT) or 0)

    # Setup output writer (use same FPS)
    fourcc = cv2.VideoWriter_fourcc(*'mp4v')
    out = cv2.VideoWriter(out_path, fourcc, src_fps, (target_w, target_h))

    # compute dot absolute position
    dot_x = int(target_w * DOT_REL_POS[0])
    dot_y = int(target_h * DOT_REL_POS[1])
    dot_pos = (dot_x, dot_y)

    prev_time = time.time()
    processed = 0
    fps_smooth = 0.0

    print(f"Processing {in_path} -> {out_path} at {src_fps:.1f} FPS, frames: {frame_count}")

    while True:
        ret, frame = cap.read()
        if not ret:
            break
        # resize for speed/consistency
        frame = cv2.resize(frame, (target_w, target_h))
        hand_info, mask = get_hand_centroid_from_frame(frame)

        # Draw target dot
        cv2.circle(frame, dot_pos, DOT_RADIUS, (255,255,255), -1)
        cv2.circle(frame, dot_pos, DOT_RADIUS, (0,0,0), 2)

        state_text = "NO HAND"
        state_color = (200,200,200)
        dist_text = "---"

        if hand_info is not None:
            cx, cy, area, contour = hand_info
            cv2.drawContours(frame, [contour], -1, (180,180,180), 2)
            cv2.circle(frame, (cx, cy), 6, (0,255,0), -1)

            dx = cx - dot_pos[0]
            dy = cy - dot_pos[1]
            dist = int(np.hypot(dx, dy))
            dist_text = f"{dist} px"
            state_text, state_color = compute_state(dist)
            cv2.line(frame, (cx, cy), dot_pos, state_color, 2)

            if state_text == "DANGER":
                cv2.putText(frame, "DANGER DANGER", (20, 60), FONT, 1.6, COLOR_DANGER, 4, cv2.LINE_AA)

        # overlay info box
        overlay = frame.copy()
        cv2.rectangle(overlay, (0, target_h-90), (target_w, target_h), (0,0,0), -1)
        alpha = 0.45
        cv2.addWeighted(overlay, alpha, frame, 1 - alpha, 0, frame)

        cv2.putText(frame, f"State: {state_text}", (16, target_h-50), FONT, 0.9, state_color, 2, cv2.LINE_AA)
        cv2.putText(frame, f"Distance: {dist_text}", (16, target_h-20), FONT, 0.7, (220,220,220), 2, cv2.LINE_AA)

        # fps calc
        cur_time = time.time()
        dt = cur_time - prev_time
        prev_time = cur_time
        fps = 1.0/dt if dt>0 else 0.0
        fps_smooth = fps_smooth*0.85 + fps*0.15 if processed>0 else fps
        cv2.putText(frame, f"FPS: {fps_smooth:.1f}", (target_w-140, 30), FONT, 0.8, (200,200,200), 2, cv2.LINE_AA)

        out.write(frame)

        processed += 1
        if processed % 50 == 0:
            print(f"Processed {processed}/{frame_count} frames, FPS: {fps_smooth:.1f}")

    cap.release()
    out.release()
    print(f"Done. Processed {processed} frames. Output saved to: {out_path}")

# Run processing
process_video(VIDEO_PATH, OUTPUT_PATH, FRAME_WIDTH, FRAME_HEIGHT)

# If in Colab, provide download link
try:
    from google.colab import files as _files
    _files.download(OUTPUT_PATH)
except Exception:
    pass
```

```python
import cv2
import numpy as np
import time

# ---------------------------------------------
# Parameters
# ---------------------------------------------
DOT_POS = (320, 240)   # Virtual point on screen
SAFE_DIST = 200        # Pixel distance thresholds
WARNING_DIST = 120
DANGER_DIST = 60

# ---------------------------------------------
# Helper: Detect hand by contour (simple skin mask)
# ---------------------------------------------
def detect_hand(frame):
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)

    # Skin color range (works for many skin tones)
    lower = np.array([0, 20, 70])
    upper = np.array([20, 255, 255])
    mask = cv2.inRange(hsv, lower, upper)

    mask = cv2.GaussianBlur(mask, (7, 7), 0)
    contours, _ = cv2.findContours(mask, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    if len(contours) == 0:
        return None, mask

    # Largest contour = hand
    c = max(contours, key=cv2.contourArea)
    if cv2.contourArea(c) < 8000:
        return None, mask

    # Find centroid
    M = cv2.moments(c)
    if M["m00"] == 0:
        return None, mask

    cx = int(M["m10"] / M["m00"])
    cy = int(M["m01"] / M["m00"])
    return (cx, cy), mask

# ---------------------------------------------
# Safety State Logic
# ---------------------------------------------
def get_state(dist):
    if dist > SAFE_DIST:
        return "SAFE", (0, 255, 0)
    elif dist > WARNING_DIST:
        return "WARNING", (0, 255, 255)
    elif dist > DANGER_DIST:
        return "DANGER", (0, 128, 255)
    else:
        return "DANGER DANGER", (0, 0, 255)

# ---------------------------------------------
# Main Program (Live Webcam)
# ---------------------------------------------
cap = cv2.VideoCapture(0)

print("Press Q to exit")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    h, w = frame.shape[:2]

    # Detect hand
    hand_pos, mask = detect_hand(frame)

    # Draw target dot
    cv2.circle(frame, DOT_POS, 10, (255, 0, 0), -1)

    if hand_pos:
        cx, cy = hand_pos
        cv2.circle(frame, (cx, cy), 10, (0, 255, 255), -1)
        cv2.line(frame, (cx, cy), DOT_POS, (255, 255, 255), 2)

        # Distance
        dist = int(np.linalg.norm(np.array([cx - DOT_POS[0], cy - DOT_POS[1]])))

        # State
        state, color = get_state(dist)
        cv2.putText(frame, state, (50, 50), cv2.FONT_HERSHEY_SIMPLEX, 1.5, color, 3)

    # Show frames
    cv2.imshow("Hand Safety System", frame)
    cv2.imshow("Mask", mask)

    # Quit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
